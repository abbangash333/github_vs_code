```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
titanic = sns.load_dataset("titanic")
```

```python
sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic)
plt.show()
```

```python
titanic.head(20)
```

```python
# coutn plot
sns.countplot(hue="class",data=titanic,x="sex")
plt.title("Plot for Counting")
plt.show()
```

### Scatter Plot

```python
g = sns.FacetGrid(titanic, row="sex", hue ="class")
g = (g.map(plt.scatter, "age", "fare", ).add_legend())
plt.show()
```
