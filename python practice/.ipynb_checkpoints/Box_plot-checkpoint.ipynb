```python
# We already discuss that we can horizontal or vertical graphs by just changing x and y value
# But we can also vertical or horizontal it by just giving orientation
# Will give an error if we have no numeric value at x (variable)
# import important library
import seaborn as sns
sns.set(style="whitegrid")
boat = sns.load_dataset("titanic")
sns.boxplot(x="age", y="fare", data=boat, orient="v")
```

```python
# But how we can give color to hue in box-plot
# we can use the following code palette={"Yes": "blue", "No": "red"}
#Because in hue we have only two value, yes or nO
import seaborn as sns
tip = sns.load_dataset("tips")
sns.set(style ="whitegrid")
sns.boxplot(x="tip", y="day", hue="smoker", data = tip, palette={"Yes": "blue", "No": "red"})
```

### If we want to add mean to the graph, then we will use the following code

```python
import seaborn as sns
import pandas as pnd
import numpy as np

boat = sns.load_dataset("titanic")
sns.boxplot(x="survived",
           y = "age", showmeans=True,
           data=boat)
```

```python
# How to change the color or image of the mean
import seaborn as sns
import pandas as pnd
import numpy as np

boat = sns.load_dataset("titanic")
sns.boxplot(x="survived",
           y = "age", showmeans=True, hue="survived",
            meanprops = {
                "marker": "+",
                "markersize": "12",
                "markeredgecolor": "red"
            },
            
           data=boat)
```

### How we can give lablel to different x and y labels

```python
import seaborn as sns
import pandas as pnd
import numpy as np
import matplotlib.pyplot as plt

boat = sns.load_dataset("titanic")
sns.boxplot(x="survived",
           y = "age", showmeans=True, hue="survived",
            meanprops = {
                "marker": "+",
                "markersize": "12",
                "markeredgecolor": "red"
            },
            
           data=boat)
plt.xlabel("How many survived"),
plt.ylabel("Age(survived)"),
plt.title("Graph of surviving people in titanic accident")
plt.show
```

```python
# We can change the size as well
import seaborn as sns
import pandas as pnd
import numpy as np
import matplotlib.pyplot as plt

boat = sns.load_dataset("titanic")
sns.boxplot(x="survived",
           y = "age", showmeans=True, hue="survived",
            meanprops = {
                "marker": "+",
                "markersize": "12",
                "markeredgecolor": "red"
            },
            
           data=boat)
# size of the text as well as font style can be changed
plt.xlabel("How many survived", size=14, weight ='bold'),
plt.ylabel("Age(survived)", size= 12,  weight = 'bold'),
plt.title("Graph of surviving people in titanic accident", size = 10, weight='bold')
plt.show
```
