```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
# Load dataset
flower = sns.load_dataset("iris")
sns.barplot(x="species", y="sepal_width", data=flower)
plt.show()
```

### How to group different individual in a group

```python
boat = sns.load_dataset("titanic")
boat
```

```python
sns.barplot(x="sex", y="alone", hue="who", data= boat)
plt.show()
```

```python
# How to set order in the above dataset
sns.barplot(x="sex", y="alone", hue="who", data= boat, order=["female", "male"])
plt.show()
```

```python
# How to change the color of the graph
sns.barplot(x="sex", y="alone", hue="who", data= boat, order=["female", "male"], color="red")
plt.show()
```

```python
sns.barplot(x="sex", y="alone", hue="who", data= boat, order=["female", "male"], color = "grey")
plt.show()
```

```python
sns.barplot(x="sex", y="alone", hue="who", data= boat, order=["female", "male"], color="blue")
plt.show()
```

```python
# How to remove error bars from the graph
sns.barplot(x="sex", y="alone", hue="who", data= boat, order=["female", "male"], color="green", ci=None)
plt.show()
```

```python
# We can use different color pallete as well
# Do search color pallete in seaborn you will get a lis
sns.barplot(x="sex", y="alone", hue="who", data= boat, order=["female", "male"], ci=None,
           palette="pastel")
plt.show()
```
