```python
# import important libraries
import seaborn as sns
import matplotlib.pyplot as plt
flower_dataset = sns.load_dataset("iris")
flower_dataset
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data= flower_dataset)
plt.title("Flowers Dataset")
# From where you want to start
# These lines will give you a limit that from where you can start your graph
plt.xlim(2)
plt.ylim(1)
plt.show()
```

## Style in LinePlots
- darkgrid
- whitegrid
- dark
- white
- ticks

```python
sns.lineplot(x="sepal_length", y="sepal_width", data= flower_dataset)
plt.title("Flowers Dataset")
# Use the below style to get all available colors
# plt.style.available
plt.style.use
```

```python
# Changing figure size
sns.lineplot(x="sepal_length", y="sepal_width", data= flower_dataset)
plt.title("Flowers Dataset")
# Use the below style to get all available colors
# plt.style.available
# We can give different style to the graph as given below
plt.figure()
```

```python
# Changing figure size
sns.lineplot(x="sepal_length", y="sepal_width", data= flower_dataset)
plt.title("Flowers Dataset")
# Use the below style to get all available colors
# plt.style.available
# We can give different style to the graph as given below
sns.set_style("dark")
sns.set_style(style=None, rc=None)
plt.show()
```

```python
# Changing figure size
sns.lineplot(x="sepal_length", y="sepal_width", data= flower_dataset)
plt.title("Flowers Dataset")
# Use the below style to get all available colors
# plt.style.available
# We can give different style to the graph as given below
sns.set_style("whitegrid")
sns.set_style(style=None, rc=None)
plt.show()
```
